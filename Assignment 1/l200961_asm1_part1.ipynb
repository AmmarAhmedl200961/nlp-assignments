Q1) Describe the class of strings matched by the following regular expressions.

- [a-zA-Z]+
  Contains one or more alphabets.Matches both capital and small letters
- [A-Z][a-z]\*
  Start with Capital letter alphabet followed by zero or more small letters
- p[aeiou]{,2}t
  Word starts with p ends with t with letters 'aeiou' in-between with 0-2 occurences.
- \d+(\.\d+)?
  starts with digit followed by one or more digits followed by zero or more pattern matching any character followed by one or more digts
- ([^aeiou][aeiou][^aeiou])\*
  starts with letter other than 'aeiou' followed by 'aeiou' and a letter not in 'aeiou' with zero or more condition
- \w+|[^\w\s]+
  Either one or more alpha numeric or not of alphanumeric and space


```python
# testing answers with nltk.re_show() function
import nltk, re, pprint

nltk.re_show('[a-zA-Z]+','This is test 123')

nltk.re_show('[A-Z][a-z]*','This IS test 123')

nltk.re_show('p[aeiou]{,2}t', 'paat pit pooot 123 abc')

nltk.re_show(r'\d+(.\d+)?','1t23 1 abc')

nltk.re_show('([^aeiou][aeiou][^aeiou])*','hat cat aei')

nltk.re_show(r'\w+|[^\w\s]+','1ab @@@')
```

```python
# Q4) Tokenize text parsed from the above url using nltk. Find all phone numbers and email addresses from
# this text using regular expressions. (Do not tokenize text otherwise email addresses will be incorrectly
# tokenized)

from selenium import webdriver
import nltk
from bs4 import BeautifulSoup
import re

# Initialize a browser driver
driver = webdriver.Firefox()

driver.get('https://www.csail.mit.edu/people/')
# last version had the javascript load up so we will be using internet archive to get the html of the page

# # Open the webpage
# driver.get('https://web.archive.org/web/20240119062816/https://www.csail.mit.edu/people/')

# Get the HTML of the page after the JavaScript has loaded
html = driver.page_source

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html, 'html.parser')

# Find all person cards
person_cards = soup.select('lib-people-result-card')

# Regular expression for extracting email addresses
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'

# Regular expression for extracting phone numbers like "253-5856"
phone_pattern = r'\b\d{3}[-.]?\d{4}\b'

for card in person_cards:
    # Extract and print email addresses and phone numbers from each card
    contact_details = card.select('lib-contact-details')
    for detail in contact_details:
        detail_html = str(detail)
        emails = nltk.regexp_tokenize(detail_html, email_pattern)
        phones = nltk.regexp_tokenize(detail_html, phone_pattern)
        print("Emails found:", list(set(emails)))
        print("Phone numbers found:", list(set(phones)))

# Close the browser
driver.quit()
```

```python
# Q5) Use the Porter Stemmer to normalize some tokenized text, calling the stemmer on each word. Do the
# same thing with the Lancaster Stemmer and see if you observe any differences
from nltk.corpus import movie_reviews

tokens = movie_reviews.words(movie_reviews.fileids()[2])
```

```python
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
```

```python
porter_list = [porter.stem(t) for t in tokens][0:20]
```

```python
lancaster_list = [lancaster.stem(t) for t in tokens][0:20]
```

```python
for i in range(20):
    if porter_list[i] != lancaster_list[i]:
        print(porter_list[i], "vs.", lancaster_list[i])
```
